## Reading in Test Data
In this section of the notebook we read in all the images and convert them to a form that is acceptable by the classifier. To accomplish this, we must first get a list of file paths in our testing directories. After this, we will read in those images, and convert them to a classifier-acceptable numerical form.

### Getting list of Image Paths
Using `listdir`, `isfile`, and `join` from the `os`, we are able to get the paths of all our testing images as a list of strings. To avoid system files we remove any non-png files from the list.

```python
symbols = [str(x) for x in range(10)]+["+","-",",","(",")",]

file_lists = {}
for symbol in symbols:
    mypath =  "img/testing/"+symbol
    file_lists[symbol] = [f for f in listdir(mypath) if isfile(join(mypath, f))]
     
    # Removing all non-png files
    for file in file_lists[symbol]:
        if(file[-4:] != ".png"):
            file_lists[symbol].remove(file)
```

### Loading in Images
Using the paths we just collected and PIL, we can load all of our test images

```python
images = {}
for symbol in symbols:
    images[symbol] = []
    for file in file_lists[symbol]:
        # Getting the full path
        path = "img/testing/"+symbol+"/"+file
        
        # Reading the image in and storing it
        image = Image.open(path)
        images[symbol].append(image.resize((45, 45)))
```

Example:

```python
plt.imshow(images["+"][0], cmap=plt.cm.binary)
plt.xticks([])
plt.yticks([])
plt.title("+")
plt.show()
```

### Converting Each Image to Mono-chrome

```python
images_mono = {}
for symbol in symbols:
    print(symbol)
    images_mono[symbol] = []
    for i in images[symbol]:
        images_mono[symbol].append(np.asarray( i, dtype="int32" ))
```

### Converting to a One-Dimensional Array

```python
# Reshaping digits
images_shaped = images_mono.copy()

for symbol in symbols:
    for image in range(len(images_shaped[symbol])):
        images_shaped[symbol][image] = images_shaped[symbol][image].reshape(-1, 45**2)
```

```python
print("first Zero: ")
images_shaped["0"][0]
```

### Converting the pixel values


```python
dfs = {}
for symbol in symbols:
    print("Math Symbol: "+symbol)
    print("Creating DF")
    dfs[symbol] = pd.DataFrame([images_shaped[symbol][i][0] for i in range(len(images_shaped[symbol]))])
#     print("Converting...")
#     dfs[symbol] = dfs[symbol].apply(lambda x : (x/32).astype(int) )
    print("")
```

### Making a scaled copy of the data
Since our calssifiers were trained on scaled data, we will need to make a copy of our data that we will scale. This will allow us to test both our non-scaled trained and scaled trained classifiers at the same time

```python
# Copying the data
sc_list = []
for symbol in symbols:
    temp = dfs[symbol].copy()
    temp["class"] = symbol
    sc_list.append(temp)
    
sc_df = pd.concat(sc_list)

# Getting the features
features = [x for x in sc_df]
features.remove("class")

# Seperating features and classes
X = np.array(sc_df[features].values)
y = sc_df["class"].values

# Scaling the features
sc = preprocessing.StandardScaler()
sc_X = sc.fit_transform(X)

# Storing the scaled data
sc_dict = {}
for symbol in symbols:
    sc_dict[symbol] = []

for i in range(len(sc_X)):
    sc_dict[y[i]].append(sc_X[i])
```

## Testing
Now that all of our test images are in the right form, we can test our classifiers

### Reading in Classifiers
In the notebook where we made our classifiers, we used the pickle library to export them to files. Here we are reading in those classifiers

```python
with open('sc_NN.pk', 'rb') as fin: # Nerual Network
    sc_clf_NN = pickle.load(fin)
with open('sc_KN.pk', 'rb') as fin: # KNeighbors
    sc_clf_KN = pickle.load(fin)
```

## Testing the Classifiers
Now that all of that is our of the way we can now train our data

```python
correct_sc_NN = 0
correct_sc_KN = 0

count = 0
misclassified = []
for symbol in symbols:
    print(f"Testing symbol: \'{symbol}\'")
    
    for sc_image in sc_dict[symbol]:
        count += 1
        sc_image = sc_image.reshape(1, -1)     

        if(sc_clf_NN.predict(sc_image)[0] == symbol):
            correct_sc_NN += 1
        if(sc_clf_KN.predict(sc_image)[0] == symbol):
            correct_sc_KN += 1
            
        if(sc_clf_NN.predict(sc_image)[0] != symbol):
            misclassified.append(sc_image)
```

```python
scores = [count, correct_sc_NN, correct_sc_KN]
labels = ["total","sc_NN","sc_KN"]
colors = ["gold", "green", "blue"]

plt.figure(figsize=(8,6))
plt.bar(labels, scores, color=colors)
m = max(scores[1:])
plt.plot([-0.5,2.5], [m,m], lw=3, color="green", ls="--", label="max")
plt.legend()
scores
```

Clearly the scaled neural network worked the best.
